```python
import pandas as pd
from sklearn import metrics
```

```python
train = pd.read_csv("./train.csv", header=None,)
```

```python
train.head()
```

```python
train.describe()
```

```python
lablel = pd.read_csv("./trainLabels.csv", header=None,)
lablel.describe()
```

```python
X = train.values
```

```python
Y = lablel.values
```

```python
from sklearn import preprocessing
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.25, random_state=42)
```

# Do Normalization

```python
from sklearn.preprocessing import StandardScaler
stds = StandardScaler()
X_train_scaled = stds.fit_transform(X_train)
X_test_scaled = stds.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf_l2_LR = LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=0.1, 
                               fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, 
                               solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, 
                               warm_start=False, n_jobs=1)
clf_l2_LR.fit(X_train_scaled,y_train)
predict_y = clf_l2_LR.predict(X_test_scaled)
print("accuracy_score",metrics.accuracy_score(y_test,predict_y))
print("f1_score",metrics.f1_score(y_test,predict_y,average='micro'))
```

```python
clf_l2_LR.coef_
```

# Build a Final Classifier

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
stds_final = StandardScaler()
X_train = stds_final.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
clf_l2_LR = LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=0.1, 
                               fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, 
                               solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, 
                               warm_start=False, n_jobs=1)
clf_l2_LR.fit(X_train,Y)
predict_y = clf_l2_LR.predict(X_train)
print("accuracy_score",metrics.accuracy_score(Y,predict_y))
print("f1_score",metrics.f1_score(Y,predict_y,average='micro'))
```

# Do Prediction

```python
test = pd.read_csv("./test.csv", header=None,)
test.describe()
```

```python
test_scaled = stds_final.transform(test)
```

```python
test_result = clf_l2_LR.predict(test_scaled)
```

```python
test_result_pd = pd.DataFrame(test_result,columns=['Solution'])
test_result_pd.index += 1 
```
