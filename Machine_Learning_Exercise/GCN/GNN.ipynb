### 從熱力學溫度是由高往低散佈
改用被影響的觀點來看會更好理解，每一個結點會被有連接(指向)的點影響，故矩陣改用 transpose of A

```python
import numpy as np
np.random.seed(168)
import networkx as nx

A = np.matrix([
    [0, 1, 0, 0],
    [0, 0, 1, 1], 
    [0, 1, 0, 0],
    [1, 0, 1, 0]],
    dtype=float
)
# 這邊 1 代表有指其它人，以有向圖來看
```

```python
# initial graph setting
AA = A.tolist()

node_size=(1000, 1000, 1000, 1000)
nodes=[0,1,2,3]
edges=[(x,y  )for x in range(4)
        for y in range(4)
            if AA[x][y] ==1 ]



G = nx.DiGraph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
pos=nx.spring_layout(G)
```

```python
def draw(node_size=(1000, 1000, 1000, 1000),nodes=[0,1,2,3], edges=[(x,y  )for x in range(4)
        for y in range(4)
            if AA[x][y] ==1 ],):
    G = nx.DiGraph()
    G.add_nodes_from(nodes)
    G.add_edges_from(edges)
    nx.draw_networkx(G,pos=pos,labels={0:0,1:1,2:2,3:3},node_size=node_size)
draw()
```

```python
# assign initial data 
H = np.ones([4,1])
H
```

```python
T = H
print(A.shape)
print(T.shape)
```

```python
draw(node_size=(T*250).reshape(4).tolist()[0])
T = A.transpose()*T
```

```python
draw(node_size=(T*250).reshape(4).tolist()[0])
T = A.transpose()*T
```
