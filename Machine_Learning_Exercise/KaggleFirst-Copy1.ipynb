```python
import pandas as pd
from sklearn import metrics
```

```python
train = pd.read_csv("./train.csv", header=None,)
```

```python
lablel = pd.read_csv("./trainLabels.csv", header=None,)
lablel.describe()
```

```python
X = train.values
```

```python
Y = lablel.values
```

```python
from sklearn import preprocessing
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.25, random_state=1)
```

# Do Normalization

```python
from sklearn.preprocessing import StandardScaler
stds = StandardScaler()
X_train_scaled = stds.fit_transform(X_train)
X_test_scaled = stds.transform(X_test)
```

# Dimession Reduction

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=40,whiten=True)
X_train_pca = pca.fit_transform(X_train_scaled)
```

```python
X_test_pca = pca.transform(X_test_scaled)
```

```python
def qq_plot(x):
    from scipy.stats import probplot
    probplot(x, dist='norm', plot=plt)
    
qq_plot(X_train_pca[:,36])
```

```python
from sklearn.linear_model import LogisticRegression
clf_l2_LR = LogisticRegression(penalty='l1', dual=False, tol=0.001, C=10, 
                               fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, 
                               solver='liblinear', max_iter=1000, multi_class='ovr', verbose=0, 
                               warm_start=False, n_jobs=1)
from sklearn import svm
# X = [[0, 0], [1, 1]]
# y = [0, 1]
# clf = svm.SVC()

clf_l2_LR = svm.SVC(C=1.5, kernel='rbf', degree=1, gamma='auto', coef0=0.0, shrinking=True, probability=False, 
                    tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, 
                    decision_function_shape='ovr', random_state=None)



clf_l2_LR.fit(X_train_scaled,y_train)
predict_y = clf_l2_LR.predict(X_test_scaled)
print("accuracy_score",metrics.accuracy_score(y_test,predict_y))
print("f1_score",metrics.f1_score(y_test,predict_y,average='micro'))

print("recall_score",metrics.recall_score(y_test,predict_y,average='micro'))
print("precision_score",metrics.precision_score(y_test,predict_y,average='micro'))
```

# Build a Final Classifier

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
stds_final = StandardScaler()
X_train = stds_final.fit_transform(X)


from sklearn import decomposition
pca = decomposition.PCA(n_components=40,whiten=True)
X_train = pca.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
# clf_l2_LR = LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=0.1, 
#                                fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, 
#                                solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, 
#                                warm_start=False, n_jobs=1)

clf_l2_LR = svm.SVC(C=1.5, kernel='rbf', degree=1, gamma='auto', coef0=0.0, shrinking=True, probability=False, 
                    tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, 
                    decision_function_shape='ovr', random_state=None)

clf_l2_LR.fit(X_train,Y)
predict_y = clf_l2_LR.predict(X_train)
print("accuracy_score",metrics.accuracy_score(Y,predict_y))
print("f1_score",metrics.f1_score(Y,predict_y,average='micro'))
```
